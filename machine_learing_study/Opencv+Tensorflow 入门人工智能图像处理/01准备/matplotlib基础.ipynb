```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x=np.linspace(0,1000,1000)
y=np.random.randint(0, 10, size=(1000, 1))

plt.plot(x,y,'b',lw=0.1) #绘制折线图 1x 2 y 3 color 4 line w
#折线 饼状 柱状
plt.show()
```

```python
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([3,4,5,1,2,4,5,6])
plt.plot(x,y,'b',lw=10) #绘制折线图 1x 2 y 3 color 4 line w

x=np.array([1,2,3,4,5,6,7,8])
y=np.array([13,14,15,11,12,14,15,16])
plt.bar(x,y,0.5,color='r',alpha=0.5) #绘制柱状图 3 占宽 5 透明度
#折线 饼状 柱状
plt.show()
```
