```python
# mnist_forward.py
# 前项网络层搭建
import tensorflow as tf

#输入维度  输出维度 中间节点设置
INPUT_NODE = 784
OUTPUT_NODE = 10
LAYER1_NODE = 500

# 权重层设置
def get_weight(shape,regularizer):
    w=tf.Variable(tf.truncated_normal(shape,stddev=0.1))
    if regularizer !=None : 
        tf.add_to_collection('losses',tf.contrib.layers.l2_regularizer(regularizer)(w))
    return w
# 偏置参数设置
def get_bias(shape):
    b = tf.Variable(tf.zeros(shape))
    return b

# 前向网络层
def forward(x,regularizer):
    # 第一层  权重层 输入维度 中间维度 正则化
    w1=get_weight([INPUT_NODE,LAYER1_NODE],regularizer)
    # 偏置参数 中间维度
    b1=get_bias([LAYER1_NODE])
    # relu激活函数 对输出结果
    y1=tf.nn.relu(tf.matmul(x,w1)+b1)
    
    # 第二层 中间维度 输出维度 正则化
    w2=get_weight([LAYER1_NODE,OUTPUT_NODE],regularizer)
    # 偏置参数 输出维度
    b2=get_bias([OUTPUT_NODE])
    # 输出结果
    y=tf.matmul(y1,w2)+b2
    # 返回最后结果
    return y

# mnist_backward.py
# 定义反向梯度求导层
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
import os

BATCH_SIZE = 200  #每一步训练的样本个数
LEARNING_RATE_BASE = 0.1  # 最开始学习率
LEARNING_RATE_DECAY = 0.99 # 学习率衰减率
REGULARIZER = 0.0001 # 正则化参数
STEPS = 50001  #总共训练步数
MOVING_AVERAGE_DECAY = 0.99 # 滑动平均衰减率
MODEL_SAVE_PATH = "./model/" # 模型保存地址
MODEL_NAME="mnist_model" # 模型保存名字

def backward(mnist):
    # x y_ 占位
    x=tf.placeholder(tf.float32,[None,INPUT_NODE])
    y_=tf.placeholder(tf.float32,[None,OUTPUT_NODE])
    # 通过前项传播程序计算y
    y=forward(x,REGULARIZER)
    # 轮数计算器 设置为不可训练
    global_step = tf.Variable(0,trainable=False)
    
    # 计算损失函数 通过交叉熵 + 正则化
    ce = tf.nn.sparse_softmax_cross_entropy_with_logits(logits=y,labels=tf.argmax(y_,1))
    cem = tf.reduce_mean(ce)
    loss = cem+tf.add_n(tf.get_collection('losses'))
    
    # 计算指数衰减学习率
    learning_rate = tf.train.exponential_decay(
        LEARNING_RATE_BASE,
        global_step,mnist.train.num_examples/BATCH_SIZE,
        LEARNING_RATE_DECAY,
        staircase=True)
    
    # 定义训练过程
    train_step = tf.train.GradientDescentOptimizer(learning_rate).minimize(loss,global_step=global_step)
    # 定义滑动平均
    ema = tf.train.ExponentialMovingAverage(MOVING_AVERAGE_DECAY,global_step)
    ema_op=ema.apply(tf.trainable_variables())
    with tf.control_dependencies([train_step,ema_op]):
        train_op=tf.no_op(name='train')
    
    #保存
    saver=tf.train.Saver()
    
    # 开始训练
    with tf.Session() as sess:
        # 初始化参数
        init_op = tf.global_variables_initializer()
        sess.run(init_op)
        # 断点继续
        ckpt = tf.train.get_checkpoint_state(MODEL_SAVE_PATH)
        if ckpt and ckpt.model_checkpoint_path:
            saver.restore(sess,ckpt.model_checkpoint_path)

        for i in range(STEPS):
            xs,ys=mnist.train.next_batch(BATCH_SIZE)
            _,loss_value,step = sess.run([train_op,loss,global_step],feed_dict={x:xs,y_:ys})
            
            if i%1000==0:
                print("After %d training step[s],loss on training batch is %g."%(step,loss_value))
                saver.save(sess,os.path.join(MODEL_SAVE_PATH,MODEL_NAME),global_step=global_step)

# mnist=input_data.read_data_sets("MNIST_data/",one_hot=True)
# backward(mnist)
```

```python
import cv2
import numpy as np
import tensorflow as tf

def restore_model(testPicArr):
    with tf.Graph().as_default() as tg:
        x=tf.placeholder(tf.float32,[None,INPUT_NODE])
        y=forward(x,None)
        preValue=tf.argmax(y,1)
        
        variable_averages=tf.train.ExponentialMovingAverage(MOVING_AVERAGE_DECAY)
        variables_to_restore = variable_averages.variables_to_restore()
        saver=tf.train.Saver(variables_to_restore)

        with tf.Session() as sess:
            ckpt = tf.train.get_checkpoint_state(MODEL_SAVE_PATH)
            if ckpt and ckpt.model_checkpoint_path:
                saver.restore(sess,ckpt.model_checkpoint_path)
                preValue=sess.run(preValue,feed_dict={x:testPicArr})
                return preValue
            else:
                print('No checkpoint file found')
                return -1

def pre_pic(picName):
    img = cv2.imread('pic/'+str(picName)+'.jpg',0)
    img=cv2.resize(img,(28,28))
    threshold = 150
    img=255-img
    # print(img)
    for i in range(28):
        for j in range(28):
            if img[i][j]<threshold:
                img[i][j]=0
            else:img[i][j]=255
    
    img=np.array(img).reshape([1,784]).astype(np.float32)/255.0
    return img


def application():
    for i in range(10):
        testPic=pre_pic(i)
        preValue=restore_model(testPic)
        print("%d prediction number is :"%i,preValue)
        

```

```python
application()
```
