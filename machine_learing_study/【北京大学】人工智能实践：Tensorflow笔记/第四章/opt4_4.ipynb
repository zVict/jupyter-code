```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties  
font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=14)  
%matplotlib inline
```

```python
w=tf.Variable(tf.constant(5,dtype=tf.float32))

loss = tf.square(w+1)

train_step = tf.train.GradientDescentOptimizer(0.2).minimize(loss)

d_y1=[]
d_y2=[]
d_x=[]

with tf.Session() as sess:
    init_op = tf.global_variables_initializer()
    sess.run(init_op)
    for i in range(40):
        sess.run(train_step)
        w_val=sess.run(w)
        loss_val=sess.run(loss)
        d_y1.append(w_val)
        d_y2.append(loss_val)
        d_x.append(i)
plt.figure()
plt.title('w_val',fontsize='large',fontweight='bold',fontproperties=font) #设置字体大小与格式
plt.xlabel("迭代次数",fontproperties=font)
plt.ylabel("w_val",fontproperties=font)
plt.plot(d_x,d_y1,'b',lw=1) #绘制折线图 1x 2 y 3 color 4 line w
plt.figure()
plt.title('loss_val',fontsize='large',fontweight='bold',fontproperties=font) #设置字体大小与格式
plt.xlabel("迭代次数",fontproperties=font)
plt.ylabel("loss_val",fontproperties=font)
plt.plot(d_x,d_y2,'b',lw=1) #绘制折线图 1x 2 y 3 color 4 line w
plt.show()
```
