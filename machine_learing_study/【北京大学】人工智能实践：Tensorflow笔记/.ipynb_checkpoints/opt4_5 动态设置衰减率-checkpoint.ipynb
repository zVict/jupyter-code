```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties  
font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=14)  
%matplotlib inline
```

```python
LEARNING_RATE_BASE=0.1  #最初学习率
LEARNING_RATE_DECAY=0.99  #学习率衰减
LEARNING_RATE_STEP=1  #喂入多少轮BATCH_SIZE后，更新一次学习率 ，一般设为:总样本/BATCH_SZIE

#运行几轮BATCH_SIZE 的计数器 ,初值为0，设为不被训练
global_step = tf.Variable(0,trainable=False)
#定义指数下降学习率
learining_rate = tf.train.exponential_decay(LEARNING_RATE_BASE,global_step,LEARNING_RATE_STEP,LEARNING_RATE_DECAY,staircase=True)

w=tf.Variable(tf.constant(5,dtype=tf.float32))

loss = tf.square(w+1)

train_step = tf.train.GradientDescentOptimizer(learining_rate).minimize(loss,global_step=global_step)

d_y1=[]
d_y2=[]
d_x=[]

with tf.Session() as sess:
    init_op = tf.global_variables_initializer()
    sess.run(init_op)
    for i in range(40):
        sess.run(train_step)
        w_val=sess.run(w)
        loss_val=sess.run(loss)
        d_y1.append(w_val)
        d_y2.append(loss_val)
        d_x.append(i)
plt.figure()
plt.title('w_val',fontsize='large',fontweight='bold',fontproperties=font) #设置字体大小与格式
plt.xlabel("迭代次数",fontproperties=font)
plt.ylabel("w_val",fontproperties=font)
plt.plot(d_x,d_y1,'b',lw=1) #绘制折线图 1x 2 y 3 color 4 line w
plt.figure()
plt.title('loss_val',fontsize='large',fontweight='bold',fontproperties=font) #设置字体大小与格式
plt.xlabel("迭代次数",fontproperties=font)
plt.ylabel("loss_val",fontproperties=font)
plt.plot(d_x,d_y2,'b',lw=1) #绘制折线图 1x 2 y 3 color 4 line w
plt.show()
```
