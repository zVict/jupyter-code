```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
class idfKMeans:
    def __init__(self, n_clusters=5, max_iter=300, tol=1e-7):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.tol = tol
    
    def k_means(self, X, n_clusters):
        n_smaples, n_features = X.shape
        self.centers_ = X[np.random.randint(0, n_smaples, self.n_clusters)]
        self.labels_ = np.zeros((n_smaples, 1))
        counter = 0
        update = True
        while(update):
            counter = counter+1
            if(counter == self.max_iter):
                break
            for i in range(n_smaples):
                x_i = X[i]
                dist = [np.linalg.norm(x_i-center) for center in self.centers_]
                self.labels_[i] = np.argmin(dist)
            update = False
            for j in range(self.n_clusters):
                index = np.where(self.labels_==j)
                if(len(index[0])==0):
                    continue
                    # raise ValueError('Please reduce number of clusters!')
                # 这里np.mean要加axis=0对列求均值
                new_center = np.mean(X[index[0]], axis=0)
                if(np.linalg.norm(self.centers_[j]-new_center) > self.tol):
                    update = True
                    self.centers_[j] = new_center
        return self.centers_, self.labels_
    
    def fit(self, X):
        self.centers_, self.labels_ = self.k_means(X, n_clusters=self.n_clusters)
        return self
                
```

```python
X = np.random.random((200, 2))
idf_clf = idfKMeans(n_clusters=10)
idf_clf.fit(X)
clf = KMeans(n_clusters=10)
clf.fit(X)
```

```python
x = X[:, 0]
y = X[:, 1]
center_x = idf_clf.centers_[:, 0]
center_y = idf_clf.centers_[:, 1]
n=np.arange(idf_clf.n_clusters)

for i,txt in enumerate(n):
    plt.annotate(txt,(center_x[i],center_y[i]))

for i in range(idf_clf.n_clusters):
    index = np.where(idf_clf.labels_==i)
    plt.scatter(x[index[0]], y[index[0]])
```

```python
for i in range(clf.n_clusters):
    index = np.where(clf.labels_==i)
    plt.scatter(x[index[0]], y[index[0]])
```
